Notebook for looking at the new molecules

```python
import pandas as pd
import numpy as np

from dock2hit.fine_tuning.process_data import read_and_process_ugi_data

df_ugi_data = read_and_process_ugi_data()
df_ugi_data.query('data_source == "new"')
```

```python
df_ugi_data.query('data_source == "new"').describe()
```

```python
df_ugi_data['IC50'].quantile(0.2)
```

IC50

```python
new_data_mean = df_ugi_data.query('data_source == "new"')['IC50'].mean()

old_data = df_ugi_data.query('data_source != "new"')
print(f'Percentage of data better than mean: {old_data.query("IC50 < @new_data_mean")["IC50"].count() / old_data.shape[0] * 100:.2f}%')
```

pIC50

```python
new_data_mean_pIC50 = df_ugi_data.query('data_source == "new"')['pIC50'].mean()

print(
    f'Percentage of data better than mean: {old_data.query("pIC50 > @new_data_mean_pIC50")["pIC50"].count() / old_data.shape[0] * 100:.2f}%')
```
