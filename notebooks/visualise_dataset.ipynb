```python
import pandas as pd 

import plotly.express as px
import molplotly
```

```python
df_cdk = pd.read_csv('../datasets/cdk2.csv')
df_cdk
```

```python
fig_cdk = px.strip(df_cdk,
                   y='cdk2_ic50',
                   color='scaffold',
                    # labels={'y_true': 'Measured Solubility'},
                    width=1000,
                    height=800)

app_cdk = molplotly.add_molecules(fig=fig_cdk,
                                    df=df_cdk,
                                  smiles_col='Smiles',
                                    title_col='mol_name',
                                    color_col='scaffold'
                                    )

app_cdk.run_server(mode='inline', port=8707, height=850)
```
